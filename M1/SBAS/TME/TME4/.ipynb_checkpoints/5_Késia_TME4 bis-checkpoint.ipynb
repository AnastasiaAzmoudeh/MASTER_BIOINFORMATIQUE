```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from matplotlib.legend_handler import HandlerLine2D
import random
import math as mt

#distribuition inicialle vecteur pi_0 S, L, B
pi0_149 = np.array([0.039, 0.033, 0.9280])
pi0_159 = np.array([0.019, 0.973, 0.0062])



#matrices 
p_149 = np.array([[0.61, 0.30, 0.09], [0.01, 0.99, 0.00], [0.01, 0.08, 0.90]])


p_159 = np.array([[0.58, 0.07, 0.35], [0.04, 0.47, 0.49], [0.01, 0.00, 0.99]])

#Normaliser les vecteurs π0.

def normalise(m):
    """
    Normalise a matrix
    input1 m: matrix
    output1 mN: normalised matrix
    """

    mN = m 
    
    for i in range (int(len(m))): 
        Somme = np.sum(m[i])
        for j in range (int(len (m[i]))):
            mN[i][j]= (m[i][j])/Somme    

    return mN

Np_149 = normalise(p_149)
print (normalise(p_149))
#[[0.61       0.3        0.09      ]
# [0.01       0.99       0.        ]
# [0.01010101 0.08080808 0.90909091]]

Np_159 = normalise(p_159)
print (normalise(p_159))
#[[0.58 0.07 0.35]
# [0.04 0.47 0.49]
# [0.01 0.00 0.99]]
```

<b>Exercice 3</b>: Déterminer les matrices de transition $P_{149}^n$ et $P_{159}^n$ pour $n$

```python
pi_zero_149 = np.array([[0.039, 0.928, 0.033]])
pi_zero_159 = np.array([[0.019, 0.0062, 0.973]])


#matrice P**n
def matTransN(P, n):
    """
    Compute the transition matrix after N iterations
    input1 P: matrix
    input2 n: number of iterations
    output Pn: transition matrix after N iterations
    """
    Pn = P
    for i in range(0,n-1): 
        Pn = Pn @ P
    return Pn
print("TEST\n",matTransN(p_149, 64), "\n\n")

#matrice P**n where n = 2,4,8,16,32,64
def printMat(P, n, pi_0, M_name):
    """
    Print all P**n matrices, where n in  {2,4,8,16,32,64, ...}  
    input1 P: matrix
    input2 n: number of iterations
    input3 pi_0: initial condition
    input4 M_name: matrix identifier
    output: None
    """
    liste_n = []
    ajout = 1 
    for i in range(n):
        ajout = ajout *2
        liste_n.append(ajout)
    print("liste_n",liste_n)
        
    for n_val in liste_n:  
        print("P",M_name,"**",n_val,"=\n",matTransN(P,n_val))
    

# print (printMat(p_149, 2, pi_zero_149, '149'))
# print("=========================================================")
# print (printMat(p_159, 2, pi_zero_159, '159'))

# on prend n = 6
print (printMat(p_149, 6, pi_zero_149, '149'))
print("=========================================================")
print (printMat(p_159, 6, pi_zero_159, '159'))
```

```python

def Pi_n(pi_0, P, n):
    """
    Compute Pi_n = Pi_0*P^n
    input1 pi_0: intial condition
    input2 P: transition matrix
    input3 n: number of iterations
    output Pi_n
    """
    
    Pi_n = pi_0 @ matTransN(P, n)

    return Pi_n 

print(pi_zero_149)
print(pi_zero_159)

Pi_149_20 = Pi_n(pi_zero_149, p_149, 20)
print("(Pi_149)**20 = ",Pi_149_20)

Pi_159_20 = Pi_n(pi_zero_159, p_159, 20)
print("(Pi_159)**20 = ",Pi_159_20)
```

<b> Exercice 5 </b>:
Ecrire une fonction pour simuler la trajectoire stochastique
$(x_t)_{t=0:n}$ d'une seule réalisation de la chaiine de Markov. 
Créer des population de $N=10,100,1000,10000$ trajectoires. Estimer 
la trajectoire moyenne de chaque population.


```python
def Trajectoire (n, Pi_0, P):
    """
    simulate the stochastic trajectory
    input1 n: number of iterations
    input2 Pi_0: initial condition
    input3 P: transtion matrix
    output Traj: list of states that simulates the stochastic trajectory ['S', 'B', 'L']
    """
    Traj = []
    
    # séquence aléatoire de n états
    nb_alea = []
    for i in range (n): 
        alea = random.random()
        nb_alea.append(alea)
#     print("ALEA ====> ", "len=",len(nb_alea),nb_alea)
    
#     normalisation de la matrice
    taille_Pi_0 = len(Pi_0)
    Pi_0_normal = np.zeros(taille_Pi_0)
    somme = np.sum(Pi_0)
#     print(somme)

    for i in range (len(Pi_0)): 
        Pi_0_normal[i]= Pi_0[i]/somme
#     print(Pi_0)
#     print("Pi_0_normal = ",Pi_0_normal)

    # création de la matrice pi' (avec les sommes cumulées croissantes)
    Pi_0_bis = np.zeros(taille_Pi_0)
#     print(Pi_0_bis)
    
    Pi_0_bis[0]= Pi_0_normal[0]
    for i in range (1,taille_Pi_0): 
        Pi_0_bis[i]= Pi_0_normal[i] + Pi_0_bis[i-1]
#         print(Pi_0_bis)
#     print("Pi_0_bis FINAL = ",Pi_0_bis)
    
    # création de la matrice P' (avec les sommes cumulées croissantes)
    taille_P = len(P)
#     print(taille_P)
    P_bis = np.zeros((taille_P, taille_P))
#     print(P_bis)
    
    for i in range(taille_P): 
        for j in range(taille_P): 
            if j == 0 : 
                P_bis[i][j]=P[i][j]
            else : 
                P_bis[i][j]=P[i][j] + P_bis[i][j-1]
#     print("P=\n",P)
#     print("P_bis = \n",P_bis)
    
    # trouver le premier état
    premier = nb_alea[0]
#     print("premier =",premier)
#     print("Pi_0_bis =", Pi_0_bis)
    if premier <= Pi_0_bis[0]:
        Traj.append("S")
    elif premier > Pi_0_bis[1]: 
        Traj.append("L")
    else: 
        Traj.append("B")
#     print ('Traj --> ',Traj)
    
    # trouver les états suivants
    for i in range(1, n): 
#         print("n=", n, "i=", i)
        if Traj[-1] == "S": 
            indice = 0
        if Traj[-1] == "B": 
            indice = 1
        if Traj[-1] == "L": 
            indice = 2
#         print("indice =",indice)

        suivant = nb_alea[i]
#         print("suivant = ",suivant)
#         print("P_bis[indice] =",P_bis[indice])
        if suivant <= P_bis[indice][0]:
            Traj.append("S")
        elif suivant > P_bis[indice][1]: 
            Traj.append("L")
        else: 
            Traj.append("B")
#         print ('Traj --> ',Traj)

    return Traj

# Test_traj = Trajectoire(10, pi0_149,Np_149)
# print("Test_ traj = ",Test_traj)

# print("=================================================================================================================")

# Test_traj = Trajectoire(10, pi0_159,Np_159)
# print("Test_ traj = ",Test_traj)

# Test_traj = Trajectoire(100, pi0_149,Np_149)
# print("Test_ traj = ",Test_traj)

print("=================================================================================================================")
Test_traj = Trajectoire(100, pi0_159,Np_159)
print("Test_ traj = ",Test_traj)


#Déterminer une trajectoire pour N population.
def Trajectoire_P(n, Pi_0, P, N):
    """
    Simulate a trajectory for N population.
    input1 n: number of iterations
    input2 Pi_0: initial condition
    input3 P: transtion matrix
    input3 N: population size
    output TrajP: trajectory
    """
    TrajP=[]
    for i in range (N): 
        traj = Trajectoire(n, Pi_0,P)
        TrajP.append(traj)
    
    return TrajP
print("=================================================================================================================")
Test_trajP = Trajectoire_P(10,pi0_159,Np_159, 5) 
print("Test_trajP ==> \n", Test_trajP)
    

#Estimer la trajectoire moyenne d'une chaine de markov pou N

def Trajectoire_moy(n, Pi_0, P, N):
    """
    Find the mean tragetoire
    input1 n: number of iterations
    input2 Pi_0: initial condition
    input3 P: transtion matrix
    input3 N: population size
    output TrajMoy: mean trajectory
    """
    
    TrajMoy = np.zeros((3,n))
    Traj_pop = Trajectoire_P(n, Pi_0, P, N) 
    
    for traj in Traj_pop: 
        L = len(traj)
#         print("traj =",traj)
        for i in range(L): 
            if traj[i]== 'S': 
                TrajMoy[0][i] +=1
            if traj[i]== 'B': 
                TrajMoy[1][i] +=1
            if traj[i]== 'L': 
                TrajMoy[2][i] +=1
                
#     print("AVANT\n", TrajMoy)         
    TrajMoy = TrajMoy/N # pour normaliser 
#     print("APRES\n", TrajMoy)

    
    
    
    
    return TrajMoy
print("=================================================================================================================")
TM = Trajectoire_moy(10,pi0_159,Np_159, 5)
print(TM)

#Visualiser graphique des trajectoires.
def plotTrajectory(t):
    """
    Plot Trajectory
    input1 t: trajectory
    """
#     print("==================== T ==================== \n", t)
    long, larg = t.shape
    axe_des_x = list(np.arange(1,larg+1))
#     print("axe_des_x", axe_des_x)
    
    graph = plt.plot(axe_des_x,  np.transpose(t),  label = " Dotplot", linestyle = 'dotted')
    plt.legend("SBL")
```

```python
plotTrajectory(Trajectoire_moy(20, pi1, p_149, 10))
```

```python
plotTrajectory(Trajectoire_moy(20, pi0_149, p_149, 10))
```

```python
#Calculer la probabilité d’une trajectoire (xt)t=0:
def Pb_Trajectoire(Traj, Pi_0, P): 
    """
    Compute the probability of a trajectory
    input1 Traj: a trajectory
    input2 Pi_0: initial condition
    input3 P: transtion matrix
    output Proba: ptobability
    """
#     print("TRAJ DE CET EXO = ", Traj)
    taille_traj = len(Traj)
#     print(taille_traj)
    avant = Traj[0]
#     print(avant)
    if avant == "S": 
        indice_avant = 0
    if avant == "B": 
        indice_avant = 1
    if avant == "L": 
        indice_avant = 2
#     print("indice_avant = ",indice_avant)
    
#     print(Pi_0)
    Proba = np.log(Pi_0[indice_avant])
            
    for i in range (1,taille_traj):
        avant = Traj[i-1]
        apres = Traj[i]
#         print("AVANT =", avant, "APRES = ", apres)
        
    # pour l'indice d'avant
        if avant == "S": 
            indice_avant = 0
        if avant == "B": 
            indice_avant = 1
        if avant == "L": 
            indice_avant = 2
        
        # pour l'indice apres
        if apres == "S": 
            indice_apres = 0
        if apres == "B": 
            indice_apres = 1
        if apres == "L": 
            indice_apres = 2
        
        Proba += np.log(P[indice_avant][indice_apres]) 

    return Proba


```

```python

def Histo (Trajs, Pi_0, P):
    """
    compute a histogramm of Trajectory probabilities
    input1 Trajs: a trajectory population
    input2 Pi_0: initial condition
    input3 P: transtion matrix  
    output pop: a list of probabilities
    """

    pop = []
    for traj in Trajs:
        proba = Pb_Trajectoire(traj, Pi_0, P)
        pop.append(proba) 

    return pop

```
