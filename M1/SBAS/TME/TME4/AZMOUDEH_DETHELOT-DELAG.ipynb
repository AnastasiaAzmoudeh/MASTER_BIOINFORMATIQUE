```python
pi_zero_149 = np.array([[0.039, 0.928, 0.033]])
pi_zero_159 = np.array([[0.019, 0.0062, 0.973]])


#matrice P**n
def matTransN(P, n):
    """
    Compute the transition matrix after N iterations
    input1 P: matrix
    input2 n: number of iterations
    output Pn: transition matrix after N iterations
    """
    Pn = P
    for i in range(0,n-1): 
        Pn = Pn @ P
    return Pn

print("matTransN pour p_149, N = 64\n",matTransN(p_149, 64), "\n\n")


#matrice P**n where n = 2,4,8,16,32,64
def printMat(P, n, pi_0, M_name):
    """
    Print all P**n matrices, where n in  {2,4,8,16,32,64, ...}  
    input1 P: matrix
    input2 n: number of iterations
    input3 pi_0: initial condition
    input4 M_name: matrix identifier
    output: None
    """
    liste_n = []
    ajout = 1 
    for i in range(n):
        ajout = ajout *2
        liste_n.append(ajout)
    print("liste_n",liste_n)
        
    for n_val in liste_n:  
        print("P",M_name,"**",n_val,"=\n",matTransN(P,n_val))

print (printMat(p_149, 2, pi_zero_149, '149'), "\n")
print (printMat(p_159, 2, pi_zero_159, '159'))

# on prend n = 6 pour avoir n inclus dans {2, 4, 8, 16, 32, 64}
# print("=============================Pour la matrice 149 ============================")
# print (printMat(p_149, 6, pi_zero_149, '149'))
# print("=============================Pour la matrice 159 ============================")
# print (printMat(p_159, 6, pi_zero_159, '159'))
```

```python

def Pi_n(pi_0, P, n):
    """
    Compute Pi_n = Pi_0*P^n
    input1 pi_0: intial condition
    input2 P: transition matrix
    input3 n: number of iterations
    output Pi_n
    """
    
    Pi_n = pi_0 @ matTransN(P, n)

    return Pi_n 


Pi_149_20 = Pi_n(pi_zero_149, p_149, 20)
print("(Pi_149)**20 = ",Pi_149_20)

Pi_159_20 = Pi_n(pi_zero_159, p_159, 20)
print("(Pi_159)**20 = ",Pi_159_20)
```

```python
def List_Pi_n(pi0, P, n):
    """
    Compute a list of  Pi_n
    input1 pi_0: intial condition
    input2 P: transition matrix
    input3 n: number of iterations
    output liste_M_trans: list
    """
    liste_Pi_n=[]

    liste_n = [] # n = {1,2,3,4,5,...}
    for i in range(n): 
        liste_n.append(i+1)
#     print(liste_n)

    for n_val in liste_n: 
        liste_Pi_n.append(Pi_n(pi0, P, n_val))
    

    return np.array(liste_Pi_n)

List_Pi_10_149 = List_Pi_n(pi_zero_149, p_149,5)
print(List_Pi_10_149)
```

<b> Exercice 5 </b>:
Ecrire une fonction pour simuler la trajectoire stochastique
$(x_t)_{t=0:n}$ d'une seule réalisation de la chaiine de Markov. 
Créer des population de $N=10,100,1000,10000$ trajectoires. Estimer 
la trajectoire moyenne de chaque population.


```python
def Trajectoire (n, Pi_0, P):
    """
    simulate the stochastic trajectory
    input1 n: number of iterations
    input2 Pi_0: initial condition
    input3 P: transtion matrix
    output Traj: list of states that simulates the stochastic trajectory ['S', 'B', 'L']
    """
    Traj = []
    
    # séquence aléatoire de n états
    nb_alea = []
    for i in range (n): 
        alea = random.random()
        nb_alea.append(alea)
#     print("ALEA ====> ", "len=",len(nb_alea),nb_alea)
    
#     normalisation de la matrice
    taille_Pi_0 = len(Pi_0)
    Pi_0_normal = np.zeros(taille_Pi_0)
    somme = np.sum(Pi_0)

    for i in range (len(Pi_0)): 
        Pi_0_normal[i]= Pi_0[i]/somme
#     print(Pi_0)
#     print("Pi_0_normal = ",Pi_0_normal)

    # création de la matrice Pi_0_bis (avec les sommes cumulées croissantes)
    Pi_0_bis = np.zeros(taille_Pi_0)
    
    Pi_0_bis[0]= Pi_0_normal[0]
    for i in range (1,taille_Pi_0): 
        Pi_0_bis[i]= Pi_0_normal[i] + Pi_0_bis[i-1]
#     print("Pi_0_bis FINAL = ",Pi_0_bis)
    
    # création de la matrice P_bis (avec les sommes cumulées croissantes)
    taille_P = len(P)
    P_bis = np.zeros((taille_P, taille_P))
    
    for i in range(taille_P): 
        for j in range(taille_P): 
            if j == 0 : 
                P_bis[i][j]=P[i][j]
            else : 
                P_bis[i][j]=P[i][j] + P_bis[i][j-1]
#     print("P_bis = \n",P_bis)
    
    # trouver le premier état
    premier = nb_alea[0]
#     print("premier =",premier)
#     print("Pi_0_bis =", Pi_0_bis)
    if premier <= Pi_0_bis[0]:
        Traj.append("S")
    elif premier > Pi_0_bis[1]: 
        Traj.append("L")
    else: 
        Traj.append("B")
#     print ('Traj --> ',Traj)
    
    # trouver les états suivants
    # on garde en mémoire l'état précédent pour ensuite se placer sur la bonne ligne de la matrice 
    for i in range(1, n): 
        if Traj[-1] == "S": 
            indice = 0
        if Traj[-1] == "B": 
            indice = 1
        if Traj[-1] == "L": 
            indice = 2

        suivant = nb_alea[i]
        if suivant <= P_bis[indice][0]:
            Traj.append("S")
        elif suivant > P_bis[indice][1]: 
            Traj.append("L")
        else: 
            Traj.append("B")
#         print ('Traj --> ',Traj)

    return Traj

Test_traj = Trajectoire(10, pi0_159,p_159)
print("Test_ traj \n ",Test_traj)




#Déterminer une trajectoire pour N population.
def Trajectoire_P(n, Pi_0, P, N):
    """
    Simulate a trajectory for N population.
    input1 n: number of iterations
    input2 Pi_0: initial condition
    input3 P: transtion matrix
    input3 N: population size
    output TrajP: trajectory
    """
    TrajP=[]
    for i in range (N): 
        traj = Trajectoire(n, Pi_0,P)
        TrajP.append(traj)
    
    return TrajP

Test_trajP = Trajectoire_P(10,pi0_159,p_159, 10) 
print("Test_trajP \n", Test_trajP)
    

    
    
#Estimer la trajectoire moyenne d'une chaine de markov pou N
def Trajectoire_moy(n, Pi_0, P, N):
    """
    Find the mean tragetoire
    input1 n: number of iterations
    input2 Pi_0: initial condition
    input3 P: transtion matrix
    input3 N: population size
    output TrajMoy: mean trajectory
    """
    
    TrajMoy = np.zeros((3,n))
    Traj_pop = Trajectoire_P(n, Pi_0, P, N) 
    
    for traj in Traj_pop: 
        L = len(traj)

        for i in range(L): 
            if traj[i]== 'S': 
                TrajMoy[0][i] +=1
            if traj[i]== 'B': 
                TrajMoy[1][i] +=1
            if traj[i]== 'L': 
                TrajMoy[2][i] +=1
                
         
    TrajMoy = TrajMoy/N # pour normaliser 



    return TrajMoy

Test_TrajMoy = Trajectoire_moy(10,pi0_159,p_159, 10)
print("Test_TrajMoy \n", Test_TrajMoy)

#Visualiser graphique des trajectoires.
def plotTrajectory(t):
    """
    Plot Trajectory
    input1 t: trajectory
    """

    long, larg = t.shape
    axe_des_x = list(np.arange(1,larg+1))

    
    graph = plt.plot(axe_des_x,  np.transpose(t),"-o",  label = " Dotplot")
    plt.legend("SBL")
```
